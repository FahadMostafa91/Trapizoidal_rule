```python
#packages 
import numpy as np
import matplotlib.pyplot as plt
#integral of f(x) from a to b by the trapezoid rule
def trapz(f,a,b,N=50):
      x = np.linspace(a,b,N+1) # N+1 points make N subintervals
      y = f(x)
      y_right = y[1:] # right endpoints # 
      y_left = y[:-1] # left endpoints
      h = (b - a)/N
      T = (h/2) * np.sum(y_right + y_left). # T= h/2*(f(x(i)+f(x(i-1)))
      return T
```

Output: Numerical Integration using Trapizoidal Rule
